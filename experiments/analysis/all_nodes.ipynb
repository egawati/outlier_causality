```python
import pandas as pd
pd.set_option('display.precision', 3)

import json
import numpy as np
import statistics

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

from ipynb.fs.full.graphics import draw_multiple_plots
```

```python
possible_root_causes_branch = {3 : ('X1',),
                        4 : ('X1','X3',),
                        5 : ('X1', 'X3', 'X4'),
                        6 : ('X1', 'X4', 'X5'),
                        7 : ('X1', 'X4', 'X5'),
                        8 : ('X1', 'X4', 'X5'),
                        9 : ('X1', 'X4', 'X5'),
                        10 : ('X1', 'X4', 'X5'),
                        11 : ('X1', 'X4', 'X5'),
                        12 : ('X1', 'X4', 'X5')}

possible_root_causes_straight = {3 : ('X1', 'X2'),
                        4 : ('X1', 'X2', 'X3'),
                        5 : ('X1', 'X2', 'X3'),
                        6 : ('X1', 'X2', 'X3'),
                        7 : ('X1', 'X2', 'X3'),
                        8 : ('X1', 'X2', 'X3'),
                        9 : ('X1', 'X2', 'X3'),
                        10 : ('X1', 'X2', 'X3'),
                        11 : ('X1', 'X2', 'X3'),
                        12 : ('X1', 'X2', 'X3'),}
```

```python
def get_performance_metrics(filepath):
    with open(filepath) as file:
        contents = file.read()
    parsed_json = json.loads(contents)
    f1 = parsed_json['accuracy_metrics']['f1']
    avg_explanation_time = parsed_json['average_explanation_time']
    return f1, avg_explanation_time

def summarise_performance_metrics(folder, vary_num_nodes, shapley_config_str, nslide_used):
    result_f1 = {}
    result_avg_exp_time = {}
    
    possible_root_causes = possible_root_causes_branch
    if 'straight' in folder:
        possible_root_causes = possible_root_causes_straight
        
    for num_nodes in vary_num_nodes:
        f1s, exp_times = [], []
        nodes = [f'X{i}' for i in range(1, num_nodes+1)]
        for root_cause in possible_root_causes[num_nodes]:
            target_node = f'X{num_nodes}'
            bname = f'{num_nodes}_nodes_{root_cause}_{target_node}'
            filename = f'performance_{bname}_{nslide_used}.json'
            if shapley_config_str:
                filepath = f'{folder}/{bname}/{shapley_config_str}/{filename}'
            else:
                filepath = f'{folder}/{bname}/{filename}'
            f1, avg_exp_time = get_performance_metrics(filepath)
            f1s.append(f1)
            exp_times.append(avg_exp_time)
        result_f1[num_nodes] = statistics.mean(f1s)
        result_avg_exp_time[num_nodes] = statistics.mean(exp_times)
    return result_f1, result_avg_exp_time
```

## Varying number of nodes of causal graph with branches

```python
causal_graph_type = 'branch'

vary_num_nodes = [i for i in range(3,11)]

alg = 'ocular'
folder = f'../{alg}/result/nodes/{causal_graph_type}'
shapley_config_str = "AUTO" 
f1_scores_ocular, avg_time_ocular = summarise_performance_metrics(folder, 
                                                    vary_num_nodes, 
                                                    shapley_config_str,
                                                    nslide_used=2)

alg = 'gcm'
folder = f'../{alg}/result/nodes/{causal_graph_type}'
shapley_config_str = "AUTO" 
f1_scores_gcm, avg_time_gcm = summarise_performance_metrics(folder, 
                                                    vary_num_nodes, 
                                                    shapley_config_str,
                                                    nslide_used=2)

alg = 'easy_rca'
folder = f'../{alg}/result/nodes/{causal_graph_type}'
shapley_config_str = None 
f1_scores_easyrca, avg_time_easyrca = summarise_performance_metrics(folder, 
                                                    vary_num_nodes, 
                                                    shapley_config_str,
                                                    nslide_used=2)

alg = 'cloudranger'
folder = f'../{alg}/result/nodes/{causal_graph_type}'
shapley_config_str = None 
f1_scores_cloudranger, avg_time_cloudranger = summarise_performance_metrics(folder, 
                                                    vary_num_nodes, 
                                                    shapley_config_str,
                                                    nslide_used=2)

alg = 'dycause'
folder = f'../{alg}/result/nodes/{causal_graph_type}'
shapley_config_str = None 
f1_scores_dycause, avg_time_dycause = summarise_performance_metrics(folder, 
                                                    vary_num_nodes, 
                                                    shapley_config_str,
                                                    nslide_used=2)

num_nodes = f1_scores_ocular.keys()
f1_ocular = f1_scores_ocular.values()
f1_gcm = f1_scores_gcm.values()
f1_easyrca = f1_scores_easyrca.values()
f1_cloudranger = f1_scores_cloudranger.values()
f1_dycause = f1_scores_dycause.values()


data_dict = {}

data_dict[0] = {'ys': (f1_cloudranger, f1_gcm, f1_easyrca, f1_dycause, f1_ocular), 
                'X'  : num_nodes,
                'ylabel' : 'F1 Score',
                'xlabel' : 'number of vertices in the causal graph',
                'lines' : ('--s','--o', '--*', '-X', '-^' ), 
                'labels' : ('CloudRanger', 'CausalRCA', 'EasyRCA', 'DyCause', 'Ocular (ours)'), 
                'colors' : ('#e28743', '#1e8532','#3377FF', '#1fdcde', '#ff005a'),
                'xticks' : None,
               'yticks' : (0.0, 1.1)}
                        
avg_exp_time_ocular = avg_time_ocular.values()
avg_exp_time_gcm = avg_time_gcm.values()
avg_exp_time_easyrca = avg_time_easyrca.values()
avg_exp_time_cloudranger = avg_time_cloudranger.values()
avg_exp_time_dycause = avg_time_dycause.values()

data_dict[1] = {'ys' : (avg_exp_time_cloudranger, avg_exp_time_gcm, avg_exp_time_easyrca, avg_exp_time_dycause, avg_exp_time_ocular),
                'X'  : num_nodes,
                'ylabel' : 'avg exp time (seconds)',
                'xlabel' : 'number of vertices in the causal graph',
                'lines' : ('--s','--o', '--*', '-X', '-^' ), 
                'labels' : ('CloudRanger', 'CausalRCA', 'EasyRCA', 'DyCause', 'Ocular (ours)'), 
                'colors' : ('#e28743', '#1e8532','#3377FF', '#1fdcde', '#ff005a'),
                'xticks' : None,
               'yticks' : None}

fig = draw_multiple_plots(data_dict, 
                    nsubplots = 2, figsize=(12,2.5), sharey=False)

pp = PdfPages(f'all_nodes_{causal_graph_type}.pdf')
pp.savefig(fig, tight_layout=True, bbox_inches = "tight")
pp.close()
```
