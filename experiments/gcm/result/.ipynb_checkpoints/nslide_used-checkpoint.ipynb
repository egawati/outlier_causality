```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
import pandas as pd
pd.set_option('display.precision', 2)
import json
import numpy as np
```

```python
def generate_table():
    nslide_used = (1, 2, 3, 4)
    exp_names = list()
    f1_scores = [list() for _ in range(len(nslide_used))]
    run_times = [list() for _ in range(len(nslide_used))]

    seeds = (0, 7, 13)
    bname = 'rand_syn_5node_X1_X5'
    for seed in seeds:
        exp_name = f'{bname}_seed{seed}'
        exp_names.append(exp_name)
        for nslide in nslide_used:
            fname = f'performance_{exp_name}_{nslide}.json'
            filepath = f'nslide_used/{fname}'
            with open(filepath) as performance_file:
                content = performance_file.read()
            performance = json.loads(content)
            f1_scores[nslide-1].append(performance['accuracy_metrics']['f1'])
            run_times[nslide-1].append(performance['run_time'])

    seeds = (3, 47, 99, 101)
    nslide_used = (1, 2, 3, 4)
    bname = 'rand_syn_5node_X2_X5'
    for seed in seeds:
        exp_name = f'{bname}_seed{seed}'
        exp_names.append(exp_name)
        for nslide in nslide_used:
            fname = f'performance_{exp_name}_{nslide}.json'
            filepath = f'nslide_used/{fname}'
            with open(filepath) as performance_file:
                content = performance_file.read()
            performance = json.loads(content)
            f1_scores[nslide-1].append(performance['accuracy_metrics']['f1'])
            run_times[nslide-1].append(performance['run_time'])

    results = {}
    results['experiment_name'] = exp_names
    for nslide in nslide_used:
        results[f'f1 score, {nslide}'] = f1_scores[nslide-1]
    for nslide in nslide_used:    
        results[f'run_time (second), {nslide}'] = run_times[nslide-1]
    df = pd.DataFrame.from_dict(results)
    #create tuples from MultiIndex
    a = df.columns.str.split(', ', expand=True).values
    #swap values in NaN and replace NAN to ''
    df.columns = pd.MultiIndex.from_tuples([('', x[0]) if pd.isnull(x[1]) else x for x in a])
    return df
```

```python
df = generate_table()
```

```python
def generate_table_for_plotting(bname = 'rand_syn_5node_X1_X5', seeds = (0, 7, 13)):
    nslide_used = (1, 2, 3, 4)
    exp_names = list()
    nslides = list()
    f1_scores = list()
    run_times = list()
    for seed in seeds:
        exp_name = f'{bname}_seed{seed}'
        for nslide in nslide_used:
            exp_names.append(exp_name)
            nslides.append(nslide)
            fname = f'performance_{exp_name}_{nslide}.json'
            filepath = f'nslide_used/{fname}'
            with open(filepath) as performance_file:
                content = performance_file.read()
            performance = json.loads(content)
            f1_scores.append(performance['accuracy_metrics']['f1'])
            run_times.append(performance['run_time'])
    results = {}
    results['experiment_name'] = exp_names
    results['nslide'] = nslides
    results['f1_score'] = f1_scores
    results['run_time'] = run_times
    df = pd.DataFrame.from_dict(results)
    return df, set(exp_names)

def plot_lines(ax, xval, ys, xlabel, ylabel, lines, labels, colors, markersize, yticks=None):
    for i in range(len(labels)):
        ax.plot(xval, ys[i], lines[i], markersize=markersize, color=colors[i])
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel)
    ax.grid(axis='x', color='0.95')
    if yticks is not None:
        ax.set_ylim(yticks)
    return ax

def draw_multiple_plots(data_dict, 
                        nsubplots, figsize, 
                        sharey=True, markersize=10):
    """
    yvals is a dictionary 
    """
    fig, axs = plt.subplots(1, nsubplots, figsize=figsize, sharey=sharey)
    for i, key in enumerate(list(data_dict.keys())):
        axs[i] = plot_lines(ax = axs[i], 
                            xval = data_dict[key]['X'], 
                            ys= data_dict[key]['ys'], 
                            xlabel = data_dict[key]['xlabel'], 
                            ylabel = data_dict[key]['ylabel'], 
                            lines = data_dict[key]['lines'], 
                            labels = data_dict[key]['labels'],
                           colors = data_dict[key]['colors'],
                           markersize=markersize,
                           yticks = data_dict[key]['yticks'])
    axs[-1].legend(labels=data_dict[key]['labels'], loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    #plt.grid(axis='x', color='0.95')
    return fig
```

```python
df, exp_names = generate_table_for_plotting(bname = 'rand_syn_5node_X1_X5', 
                                            seeds = (0, 7, 13))
exp_names
data_dict = {}
exp_names = list(exp_names)
nslide_used = (1, 2, 3, 4)
pos = [f'{i}' for i in nslide_used]

ydata_f1 = list()
for exp_name in exp_names:
    ydata_f1.append(df.loc[df['experiment_name'] == exp_name, 'f1_score'])

data_dict[0] = {'ys': ydata_f1,
                'X'  : pos,
                'ylabel' : 'F1 Score',
                'xlabel' : 'number of slides used to explain each outlier',
                'lines' : ('--o', '-^', ':D', '--s', '-.X', ':p', '-.<'), 
                'labels' : exp_names,
                'colors' : ('#1e8532', '#ff005a', '#851e3e', '#4d648d', '#ff8b94', '#FFC20A', '#0C7BDC'),
                'yticks' : (0.6, 1.0)}

ydata_time = list()
for exp_name in exp_names:
    ydata_time.append(df.loc[df['experiment_name'] == exp_name, 'run_time'])

data_dict[1] = {'ys' : ydata_time,
                'X'  : pos,
                'ylabel' : 'running time (seconds)',
                'xlabel' : 'number of slides used to explain each outlier',
                'lines' : ('--o', '-^', ':D', '--s', '-.X', ':p', '-.<'), 
                'labels' : exp_names,
                'colors' : ('#1e8532', '#ff005a', '#851e3e', '#4d648d', '#ff8b94', '#FFC20A', '#0C7BDC'),
                'yticks' : (175, 180)}

fig = draw_multiple_plots(data_dict, 
                    nsubplots = 2, figsize=(12,3), sharey=False)
pp = PdfPages(f'nslide_V1.pdf')
pp.savefig(fig, tight_layout=True, bbox_inches = "tight")
pp.close()
```
