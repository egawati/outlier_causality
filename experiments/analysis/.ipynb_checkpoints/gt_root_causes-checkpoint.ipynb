```python
import pandas as pd
pd.set_option('display.precision', 3)

import json
import numpy as np
import scipy.stats as stats

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

from ipynb.fs.full.graphics import draw_multiple_plots
```

```python
def get_performance_metrics(filepath):
    with open(filepath) as file:
        contents = file.read()
    parsed_json = json.loads(contents)
    f1 = parsed_json['accuracy_metrics']['f1']
    avg_explanation_time = parsed_json['average_explanation_time']
    return f1, avg_explanation_time

def summarise_performance_metrics(root_causes, 
                                  random_seed =13,
                                  num_nodes=6, 
                                  folder=f'rcs/branch', 
                                  nslide_used=2,
                                 ):
    result_f1 = {}
    result_avg_exp_time = {}
    target_node = f'X{num_nodes}'
    for root_cause_gt in root_causes:
        bname = f'{num_nodes}_nodes_{root_cause_gt}_{target_node}_s{random_seed}'
        filename = f'performance_{bname}_{nslide_used}.json'
        filepath = f'{folder}/{bname}/{filename}'
        f1, avg_exp_time = get_performance_metrics(filepath)
        result_f1[root_cause_gt] = f1
        result_avg_exp_time[root_cause_gt] = avg_exp_time
    return result_f1, result_avg_exp_time
```

```python
causal_graph_type = 'branch'
root_causes = ('X1', 'X3', 'X4', 'X5')
random_seed = 13
alg = 'ocular'
folder = f'../{alg}/result/rcs/{causal_graph_type}'
shapley_config_str = "AUTO" 
f1_scores_ocular, avg_time_ocular = summarise_performance_metrics(root_causes, 
                                  random_seed =random_seed,
                                  num_nodes=6, 
                                  folder=folder, 
                                  nslide_used=2,
                                 )
alg = 'gcm'
folder = f'../{alg}/result/rcs/{causal_graph_type}'
shapley_config_str = "AUTO" 
f1_scores_gcm, avg_time_gcm = summarise_performance_metrics(root_causes, 
                                  random_seed =random_seed,
                                  num_nodes=6, 
                                  folder=folder, 
                                  nslide_used=2,
                                 )

rates_str = list(f1_scores_ocular.keys())
rates_str = [f'{i}' for i in rates_str]
f1_ocular = f1_scores_ocular.values()
f1_gcm = f1_scores_gcm.values()

data_dict = {}

data_dict[0] = {'ys': (f1_gcm, f1_ocular), 
                'X'  : rates_str,
                'ylabel' : 'F1 Score',
                'xlabel' : 'root cause',
                'lines' : ('--o', '-^', ), 
                'labels' : ('CausalRCA', 'Ocular'), 
                'colors' : ('#1e8532', '#ff005a', ),
                'xticks' : None,
               'yticks' : (0.4, 1.1)}
                        
avg_exp_time_ocular = avg_time_ocular.values()
avg_exp_time_gcm = avg_time_gcm.values()

data_dict[1] = {'ys' : (avg_exp_time_gcm, avg_exp_time_ocular),
                'X'  : rates_str,
                'ylabel' : 'avg exp time (seconds)',
                'xlabel' : 'root cause',
                'lines' : ('--o', '-^', ), 
                'labels' : ('CausalRCA', 'Ocular'), 
                'colors' : ('#1e8532', '#ff005a', ),
                'xticks' : None,
               'yticks' : None}

fig = draw_multiple_plots(data_dict, 
                    nsubplots = 2, figsize=(12,1.5), sharey=False)

pp = PdfPages(f'vary_root_causes_{causal_graph_type}_r{random_seed}.pdf')
pp.savefig(fig, tight_layout=True, bbox_inches = "tight")
pp.close()
```
